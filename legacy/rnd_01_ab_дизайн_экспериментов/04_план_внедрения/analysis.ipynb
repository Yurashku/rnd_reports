# RnD 04: операционный шаблон внедрения
Ноутбук формирует компактный чек-лист шагов перед запуском эксперимента.

```python
steps = [
    'Зафиксировать primary и secondary метрики',
    'Выбрать тип коррекции множественных проверок',
    'Зафиксировать правило приемки сплита (SMD + лимит попыток)',
    'Подготовить шаблон итогового отчёта со ссылками на код'
]
for i, step in enumerate(steps, start=1):
    print(f'{i}. {step}')
```

## Что важно
Чек-лист должен быть согласован до просмотра результатов эксперимента.
