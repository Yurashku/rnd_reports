# RnD 02: корректировки для финальных метрик
Ниже — компактный пример Bonferroni и Benjamini–Hochberg.

```python
alpha = 0.05
pvals = [0.003, 0.018, 0.031, 0.22, 0.41]
m = len(pvals)

bonf_threshold = alpha / m
print(f'Порог Bonferroni: {bonf_threshold:.4f}')
print('Значимы (Bonferroni):', [p for p in pvals if p <= bonf_threshold])

# BH-процедура
sorted_p = sorted((p, i) for i, p in enumerate(pvals, start=1))
passed = []
for rank, (p, idx) in enumerate(sorted_p, start=1):
    crit = alpha * rank / m
    if p <= crit:
        passed.append((idx, p, crit))

print('Кандидаты BH (индекс, p, порог):')
for row in passed:
    print(row)
```

## Комментарий
Bonferroni более консервативен, BH обычно даёт больше мощности при контроле FDR.
