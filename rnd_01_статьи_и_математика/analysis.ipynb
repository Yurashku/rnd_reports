# RnD 01: теоретическая база
Этот ноутбук поддерживает математическую часть и показывает, как быстро растёт риск ложных находок при множественных проверках.

```python
alpha = 0.05
for m in [1, 5, 10, 20, 50]:
    p_any = 1 - (1-alpha)**m
    bar = '█' * int(p_any * 40)
    print(f'm={m:2d} | P(хотя бы одно ложное срабатывание)={p_any:0.3f} {bar}')
```

## Вывод
Даже при умеренном числе метрик вероятность хотя бы одной ложной значимости быстро растёт, поэтому в финальном анализе нужны корректировки.
