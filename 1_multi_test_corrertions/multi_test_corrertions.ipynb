# Multiple testing corrections demo
Мини-демонстрация корректировок p-value для финальных метрик.

```python
from math import comb

alpha = 0.05
m = 5
print(f'Bonferroni threshold for {m} tests: {alpha/m:.4f}')
pvals = [0.01, 0.03, 0.2, 0.8, 0.04]
print('Raw p-values:', pvals)
print('Rejected with Bonferroni:', [p < alpha/m for p in pvals])
```
