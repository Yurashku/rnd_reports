# RnD 03: рерандомизация и баланс
Проверяем, как меняется вероятность ложного сигнала при росте числа ковариат.

```python
alpha = 0.05
for m in [5, 10, 20, 50, 100]:
    p_any = 1 - (1-alpha)**m
    print(f'ковариат={m:3d} -> P(ложный сигнал хотя бы в одной)={p_any:.3f}')
```

## Практическая рекомендация
Для приемки сплита лучше использовать заранее зафиксированный критерий дисбаланса (например, max |SMD|), а не только серию p-value проверок.
